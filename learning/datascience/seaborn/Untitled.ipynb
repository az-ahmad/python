```python
# seaborn is built on top of matplotlib
# conda install seaborn
# pip install seaborn

import seaborn as sns
%matplotlib inline

# if not using IPythong notebooks, import matplotlib.pyplot as plt , then use plt.show() command to present plots

ds1 = sns.load_dataset('tips')
ds1.head()
```

```python
# DISTRIBUTION PLOTS
```

```python
# displot
# the distplot can show distribuiton of a univariate set of observations

sns.distplot(ds1['total_bill'])

# able to remove the KDE with kde=False
```

```python
# jointplot
# the jointplot joins two distplots(two columns from the dataset you want to compare)
sns.jointplot(x='total_bill',y='tip',data=ds1)
```

```python
# pairplot
# the pairplot plots pairwise relationships across the dataset, a simplified version of PairGrid
# can compare categorical with the use of hue=''
sns.pairplot(data=ds1,hue='sex')
```

```python
# rugplots
# rugplots show points on the plot for every measure on the univariate distribution
sns.rugplot(ds1['total_bill'])
```

```python
# kde plots
# kernal density estimation plots show a normal distribution curve over all the observations
sns.kdeplot(ds1['tip'])
```

```python
# CATEGORICAL PLOTS
```

```python
# Bar Plot
# bar plots are a simple way of showing one categorical column against a numerical, usually showing the mean unless changed

sns.barplot(x='sex',y='tip',data=ds1)
```

```python
# Count Plot
# count plots have a categorical observation on x-axis and its count on the y-axis
sns.countplot(x='sex',data=ds1)
```

```python
# Box plots
# box plots will show quartiles and show any outliers. Can compare observations in another category with hue=''
sns.boxplot(x='sex',y='tip',data=ds1,hue='smoker')
```

```python
# Violin plot
# a violin plot is essentially a box plot with a kernel density estimation instead of a box
sns.violinplot(x='sex',y='tip',data=ds1,hue='smoker',split=True)
```

```python
# Scatter Plot
# a scatter plot shows the observation points of a numerical and a catergorical set of values
sns.scatterplot(x='sex',y='tip',data=ds1,hue='smoker')
```

```python
# Swarm Plot
# essentially a scatter plot but the points do not stack over each other, easier to see all the points
sns.swarmplot(x='sex',y='tip',data=ds1,hue='smoker',split=True)
```

```python
# Cat Plot
#This function provides access to several axes-level functions that
# show the relationship between a numerical and one or more categorical
# variables using one of several visual representations.

sns.catplot(x='sex',y='tip',data=ds1,kind='box')
```

```python
#MATRIX PLOTS
# for matrix plots, both indexes must represent something, cannot use default indexing 0,1,2 etc..
```

```python
# Heatmaps
# Heat maps can indicate correlation data. A simple way to do so is the following:
sns.heatmap(ds1.corr())
```

```python
# Cluster maps
# Cluster maps work like heat maps but group together the most nearest categories based on the results
# Use pivot_table(index='',columns='',values='')
sns.clustermap(ds1.pivot_table(index='sex',columns='smoker',values='tip'))
```

```python
# Grid Plots
# Grid plots can show multiple relatable plots on a grid. In most instances the grid has to be initialised first.
grid1 = sns.PairGrid(ds1)
import matplotlib.pyplot as plt
# takes in data values as default, can compare categroical with hue=''

#can change the upper/diagonal/lower plots
grid1.map_upper(plt.scatter)
grid1.map_diag(sns.kdeplot)
grid1.map_lower(plt.bar)
```

```python
# Can use simplified sns.pairplot()
sns.pairplot(ds1)
```

```python
# Facet Grid
# Create grid plots and select the type of plot you want
# initialise the grid with the row and column comparisons
grid2 = sns.FacetGrid(ds1,row='smoker',col='sex')
#plot to the grid with the data values you want to look at
grid2.map(plt.hist,'tip')
```

```python
# Regression Plot
# display linear models with regression. takes in numerical categorical data for x and y, categorize with hue and row=''col=''
sns.lmplot(x='tip',y='total_bill',data=ds1,hue='sex',col='sex',row='smoker')
```
